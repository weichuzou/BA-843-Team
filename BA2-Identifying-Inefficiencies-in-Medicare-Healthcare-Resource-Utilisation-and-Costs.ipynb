## 1. Introduction

### 1.1 Motivation

One of the biggest and most intricate sectors in the world, the healthcare sector has a vast number of players and moving elements. In the US, millions of people with needs are covered by the government-funded Medicare insurance program. In order to detect potential inefficiencies and consider solutions for lowering healthcare costs while preserving the quality of treatment, it is critical to examine how hospital resources are used and the cost of care for Medicare inpatients and outpatients.

### 1.2 Project Goal

Through this project, our team aims to collect and analyze data on various healthcare utilization and cost metrics for Medicare inpatients and outpatients across different healthcare settings, such as hospitals, nursing facilities, home health agencies, and hospice providers. By examining this data, we may spot patterns and trends in healthcare spending and utilization and potential improvement areas. The ultimate purpose of this project is to assist decision-makers in the healthcare industry in better resource allocation and enhancement of the quality and cost-effectiveness of Medicare beneficiary healthcare.

### 1.3 Dataset Information

#### Checking null values

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum

# Check for null values in each column
null_counts = hospital_general_info.select([sum(col(c).isNull().cast("int")).alias(c) for c in hospital_general_info.columns])
null_counts.show(vertical=True, truncate=False)
```

We observed that meets_criteria_for_promoting_interoperability_of_ehrs has 1450 null values, hospital_overall_rating_footnote has 3270 null values, mortaility_group_footnote has 3963 null value, safety_group_footnote has 3555 null values, readmission_measures_footnote 4353 null values, patient_experience_measures_footnote has 3291 null value, timely_and_effective_care_measures_footnote has 4407 null values.

```python
# Total number of rows in the DataFrame
total_rows = hospital_general_info.count()

# Calculate the percentage of missing values
for column in hospital_general_info.columns:
    missing_count = hospital_general_info.filter(col(column).isNull()).count()
    missing_percentage = (missing_count / total_rows) * 100
    
    print(f"{column}: {missing_percentage:.2f}% missing data")

# To drop columns with >= 50% missing data
columns_to_drop = []
for column in hospital_general_info.columns:
    missing_count = hospital_general_info.filter(col(column).isNull()).count()
    missing_percentage = (missing_count / total_rows) * 100

    if missing_percentage >= 50:
        columns_to_drop.append(column)

hospital_general_info = hospital_general_info.drop(*columns_to_drop)
```

```python
#Drop all the rows with null value
hospital_general_info = hospital_general_info.drop("meets_criteria_for_promoting_interoperability_of_ehrs")
hospital_general_info = hospital_general_info.drop("hospital_overall_rating_footnote")
hospital_general_info = hospital_general_info.drop("mortaility_group_footnote")
hospital_general_info = hospital_general_info.drop("safety_group_footnote")
hospital_general_info = hospital_general_info.drop("readmission_measures_footnote")
hospital_general_info = hospital_general_info.drop("patient_experience_measures_footnote")
hospital_general_info = hospital_general_info.drop("timely_and_effective_care_measures_footnote")
```

For meets_criteria_for_promoting_interoperability_of_ehrs column, it only has 27.17% missing data, which has been dropped as it has been not been included in the analysis. Also, we observed that "hospital_overall_rating_footnote", "mortaility_group_footnote", "safety_group_footnote", "readmission_measures_footnote", "patient_experience_measures_footnote", and "timely_and_effective_care_measures_footnote" all have over 50% missing data, so we dropped it.

#### 4.3.2 What are the most common APC codes billed by outpatient providers in each hospital referral region?

```python
from pyspark.sql.functions import desc, rank
from pyspark.sql.window import Window

# get the top 10 most common APC codes for each hospital referral region
w = Window.partitionBy("hospital_referral_region").orderBy(desc("count"))

# group by hospital referral region and apc, and count the number of providers for each apc
top_apc_per_region = outpatient_all.groupBy("hospital_referral_region", "apc").count()

# rank the APC codes by their frequency of occurrence within each referral region
ranked_apc_per_region = top_apc_per_region.withColumn("rank", rank().over(w)).filter("rank <= 10")

# join the result with the provider_name and average_total_payments to get the provider name and average total payments for each apc code
result = ranked_apc_per_region.join(outpatient_all.select("provider_id", "provider_name", "hospital_referral_region", "apc", "average_total_payments"),
                                     ["apc", "hospital_referral_region"]).select("apc", "provider_name",
                                                                                  "hospital_referral_region", "average_total_payments")

# show the top 10 most common APC codes and their corresponding providers and average total payments for each hospital referral region
result.show(10, truncate=False)
```

We can observe from the supplied output that there are considerable differences between the average total payments for each APC code. APC code 0013 (Level II Debridement & Destruction), for instance, has a minimumaverage total payment of 74.0 dollars, whereas APC code 0207 (Level III Nerve Injections) has a substantially greater average total payment of $631.0. This shows that the average total payments can be significantly impacted by the complexity and expense of the medical procedure linked with the APC code.

We can also see that, for the listed APC codes, the Holyoke Medical Center seems to be the most popular outpatient provider in the MA - Springfield hospital referral zone. ### 5.3.1. Linear Regression Model for Predicting Yearly Average Total Payments

```python
from pyspark.sql.functions import *
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression


# Prepare the data by selecting only the relevant columns and removing any missing or invalid values
payments_df = merged_table.select('year', 'average_total_payments').filter(col('average_total_payments').isNotNull())

# Perform exploratory data analysis to understand the trends and patterns in average total payments over time
payments_df.groupBy('year').agg(avg('average_total_payments')).orderBy('year').show()

# Create a new feature that calculates the average total payments for each year
yearly_payments_df = payments_df.groupBy('year').agg(avg('average_total_payments').alias('yearly_average_total_payments'))

```

```python
# Create a feature vector that includes the yearly average total payments
assembler = VectorAssembler(inputCols=['year'], outputCol='features')
yearly_payments_df = assembler.transform(yearly_payments_df).select('features', 'yearly_average_total_payments')

# Split the data into training and test sets
training_data, test_data = yearly_payments_df.randomSplit([0.7, 0.3], seed=42)

# Train a linear regression model on the training data
lr = LinearRegression(featuresCol='features', labelCol='yearly_average_total_payments')
lr_model = lr.fit(training_data)

# Evaluate the performance of the model on the test data
predictions = lr_model.transform(test_data)
mse = predictions.agg(mean((col('yearly_average_total_payments') - col('prediction'))**2)).collect()[0][0]
print(f"Mean Squared Error: {mse}")

# Use the model to make predictions for future years
future_years = [(2024,), (2025,), (2026,), (2027,), (2028,)]
future_years_df = spark.createDataFrame(future_years, ['year'])
future_years_df = assembler.transform(future_years_df).select('features')
future_predictions = lr_model.transform(future_years_df)
future_predictions.show()
```

In this code, we are performing simple linear regression to predict the average total payments for future years based on the historical data of average total payments over time. We start by selecting the relevant columns and removing any missing or invalid values. Then, we perform exploratory data analysis to understand the trends and patterns in average total payments over time.

Next, we create a new feature that calculates the average total payments for each year and create a feature vector that includes the yearly average total payments. We split the data into training and test sets and train a linear regression model on the training data.

We evaluate the performance of the model on the test data by computing the Mean Squared Error (MSE), which is a measure of the difference between the predicted and actual values. We then use the trained model to make predictions for future years, and display the predicted values.

Comparing the MSE and the predicted values for future years, we can see that the model performs reasonably well on the test data, with an MSE of 1909944.0338056427. The predicted values for future years show a clear upward trend, indicating an increase in the average total payments over time.

Although we can make predictions using only the year variable, the resulting Mean Squared Error (MSE) is quite high. This suggests that using only one feature can have a significant impact on the accuracy of the predictions. Therefore, it may be necessary to include additional features in the model in order to improve its performance.
